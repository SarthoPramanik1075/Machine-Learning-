```python
import numpy as np
arr1=np.array([1,2,3,4,5])
print(type(arr1))
print(arr1.ndim)

arr2=np.array([[1,2,3],
              [4,5,6]
              ])
print(arr2.ndim)

arr3=np.array([
    [ [1,2,3],
      [4,5,6] ],
    [ [6,7,8],
      [9,1,2] ]
])
print(arr3.ndim)
```

Array Attributesa

```python
#dimention
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

#Shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

#Data type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)

#size
print(arr1.size)
print(arr2.size)
print(arr3.size)

#selecting a data type for an array
arr=np.array([1,2,3], dtype=np.int8)
print(arr.dtype)
arr=arr.astype(np.int32)
print(arr.dtype)
```

Indexing and Slicing

```python
print(arr1)
print(arr1[2])

arr1[2]=100
print(arr1)

arr1_mod=arr1[1:4]
print(arr1_mod)
```

```python
print(arr2)

print(arr2[1][2])

row_0=arr2[0:1,]
row_1=arr2[1:]
print(row_0)
print(row_1)
```

```python
arr1_mod=arr1[1:4].copy()
arr1_mod[2]=200
print(arr1_mod)
print(arr1)
```
