```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({'h':[150,160,170,180,300],
                 'w':[50,60,70,80,200]})
df
```

```python
md=df.median()
q1=df.quantile(.25)
q3=df.quantile(.75)
q1
q3
iqr=q3-q1
md, q1, q3, iqr
```

```python
ct=df-md
ct
```

```python
r= ct/iqr
r
```

```python
r.plot(kind='bar')
```
