```python
hours=[2,3,4,1,5,7,3]
days=[1,2,3,4,5,6,7]
plt.plot(days, hours)
```

```python
df=pd.read_csv("Downloads/enrollment_data.csv")
df
```

```python
plt.xlabel('YEARS')
plt.ylabel('Count on Enrollment')
plt.title('Student Enrollment over 8 years')
plt.plot(df["Year"],df["Programming"])
plt.plot(df["Year"],df["Digital Marketing"])
```

```python
plt.xlabel('YEARS')
plt.ylabel('Count on Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df["Year"],df["Programming"], label='Programming')
plt.plot(df["Year"],df["Digital Marketing"],label='Digital Marketing')
plt.legend()
```

```python
plt.xlabel('YEARS')
plt.ylabel('Count on Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df["Year"],df["Programming"], label='Programming',linestyle='dashed',linewidth=3, marker='o', markersize=7)
plt.plot(df["Year"],df["Digital Marketing"],label='Digital Marketing')
# plt.legend(loc='lower right')
plt.legend(loc='best')
plt.grid()
plt.show()
```
