```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({'h':[150,175,170,180,190],
                 'w':[50,60,70,80,90]})
df
```

```python
mn=df.min()
mx=df.max()
rg=mx-mn
mn, mx, rg
```

```python
ss=df-mn
ss.head()
```

```python
mm=ss/rg
mm.round(2)
```

```python
mm.plot(kind='bar')
```
