```python
 from functools import reduce
with open("./sample_data/text.txt","r") as file:
  content=file.readlines()
  # koita line
  line=len(content)
/

  #word count
  num_word=map(lambda x: len(x.split()),content)
  print(num_word)
  total_word=reduce(lambda x,y: x+y, num_word)
  print(total_word)

  #cleaning process
  content=list(map(str.strip,content))
  print(content)

  #space delete
  content=list(map(lambda x: x.replace(" ",""), content))
  print(content)

  #num_character

  num_char=list(map(lambda x: len(x), content))
  print(num_char)
  total_num_char=reduce(lambda x,y: x+y, num_char)
  print(total_num_char)

  with open("./sample_data/count.txt","w") as file:
    file.write(f"Total line: {line}\nTotal num of word:{total_word}\ntotal character: {total_num_char}")
```
